```python
import os
import pandas as pd
```

```python
excel_path = "ki_price_plans_total_business.xlsx"
sheet_name = "bartotq"  # your sheet name

# Read raw sheet without headers
raw = pd.read_excel(excel_path, sheet_name=sheet_name, header=None)

# Row 2 (index 2) = quarters, from column 3 onwards
quarters = raw.iloc[2, 3:]

# Row 3 (index 3) = values, from column 3 onwards
values = raw.iloc[3, 3:]

# Build tidy dataframe
df = pd.DataFrame({"quarter": quarters.values, "value": values.values})

# Make sure values are numeric
df["value"] = pd.to_numeric(df["value"], errors="coerce")

# Save clean CSV
out_name = "ki_price_plans_total_business_long_clean.csv"
df.to_csv(out_name, index=False)

print("Saved:", out_name)
print(df.head())
```

```python
excel_path = "ki_inflation_expectations_households.xlsx"

base = os.path.splitext(excel_path)[0]

xlsx = pd.ExcelFile(excel_path)

for sheet in xlsx.sheet_names:
    df = pd.read_excel(excel_path, sheet_name=sheet)

    csv_name = f"{base}_{sheet}.csv"

    df.to_csv(csv_name, index=False)
    print(f"Saved: {csv_name}")
```
